## Observations and Insights 

DATA OBSERVATIONS:

1) The correlation coefficient of 0.84 between mouse weight and tumor volume indicates a positive relationship between the two variables in the Capomulin drug regimen. 

2) In comparison to the other drugs, Capomulin had one of the smallest average tumor volume sizes. Both Capomulin and Ramicane had mean tumor volumes below 40. 

3) The box plots show that only Infubinol had an outlier at 34. The drug's middle range is much higher and since there was only extreme value, it does not really impact the dataset. Since it doesn't change the results, it would be ok to drop the outlier as long as a note is made.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_study_data = pd.merge(mouse_metadata, study_results, how='outer', on='Mouse ID')

# Display the data table for preview
combined_study_data.head()
```

```python
# Checking the number of mice.
len(combined_study_data["Mouse ID"].value_counts())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_rows = combined_study_data[combined_study_data.duplicated(["Mouse ID", "Timepoint"])]
duplicate_rows["Mouse ID"].value_counts()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_rows
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_combined_df = combined_study_data.drop_duplicates(subset=["Mouse ID", "Timepoint"])
clean_combined_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
clean_combined_df["Mouse ID"].value_counts()
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method is the most straighforward, creating multiple series and putting them all together at the end.

mean_group_df = clean_combined_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()
grouped_summary_stats_df = pd.DataFrame(mean_group_df)
grouped_summary_stats_df = grouped_summary_stats_df.rename(columns={"Tumor Volume (mm3)": "Mean"})

median_group_df = clean_combined_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].median()
grouped_summary_stats_df["Median"] = median_group_df

variance_group_df = clean_combined_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()
grouped_summary_stats_df["Variance"] = variance_group_df

std_group_df = clean_combined_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()
grouped_summary_stats_df["Standard Deviation"] = std_group_df

sem_df = clean_combined_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()
grouped_summary_stats_df["Standard Error of Means"] = sem_df

grouped_summary_stats_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method produces everything in a single groupby function
group_sum_stats_2 = clean_combined_df.groupby("Drug Regimen").agg({"Tumor Volume (mm3)": ["mean", "median", "var", "std", "sem"]})
group_sum_stats_2
```

```python
# Generate a bar plot showing the total number of mice for each treatment throughout the course of the study using pyplot.
mice_count = clean_combined_df["Drug Regimen"].value_counts()
mice_row = clean_combined_df["Drug Regimen"].unique()
x_axis = np.arange(len(grouped_summary_stats_df))
tick_locations = [value for value in x_axis]

plt.xlim(0, len(x_axis))
plt.ylim(0, max(mice_count)+10)

plt.figure(figsize=(9,3))
plt.bar(x_axis, mice_count, align="center")
plt.xticks(tick_locations, mice_row)

plt.title("Number of Mice per drug treatment")
plt.xlabel("Drug Regimen")
plt.ylabel("Mice Count")
plt.tight_layout()
plt.savefig("Images/Py_Bar_Chart.png")
```

# QUARTILES

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Start by getting the last (greatest) timepoint for each mouse

last_tp = clean_combined_df.groupby(["Mouse ID"])["Timepoint"].max()
last_tp = last_tp.reset_index()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
tp_merge = last_tp.merge(clean_combined_df, on = ["Mouse ID", "Timepoint"], how= "left")
tp_merge
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tum_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

for i in treatments:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    tum_vol = tp_merge.loc[tp_merge["Drug Regimen"]== i, "Tumor Volume (mm3)"]
    
    # add subset 
    tum_vol_data.append(tum_vol)
    
    # Determine outliers using upper and lower bounds
    quartiles_info = tum_vol.quantile([.25,.5,.75])
    lower_range = quartiles_info[0.25]
    upper_range = quartiles_info[0.75]
    middle_range = upper_range - lower_range
    
    lower_bound = lower_range - (1.5*middle_range)
    upper_bound = upper_range + (1.5*middle_range)
    
    find_outliers = tum_vol.loc[(tum_vol < lower_bound) | (tum_vol > upper_bound)]
    print(f"{i} outliers {find_outliers}")
    
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
blue_out  = dict(markerfacecolor = 'blue', markersize=12)
fig1, ax1 = plt.subplots()
plt.boxplot(tum_vol_data, labels = treatments, flierprops = blue_out)
ax1.set_title('Tumor Volume across Drug Regimens of Interest')
ax1.set_ylabel('Tumor Volume')

plt.show()

plt.savefig("Images/BoxPlot.png")
```

```python
plt.clf()
plt.cla()
plt.close()
```

# Line and Scatter Plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin

cap_mice = clean_combined_df.loc[clean_combined_df["Drug Regimen"] == "Capomulin", :]
cap_mice_x = clean_combined_df.loc[clean_combined_df["Mouse ID"] == "s185", :]
cap_mice_x

x_axis = np.array([0,5,10,15,20,25,30,35,40,45])


plt.plot(x_axis, cap_mice_x["Tumor Volume (mm3)"], color="green", label="Timepoint")

plt.title("Mouse s185's Tumor Volume over Time")
plt.xlabel("Timepoint")
plt.ylabel("Tumor Volume")
plt.legend()

# Print our chart to the screen
plt.show()
plt.savefig("Images/Line_chart.png")
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
cap_mice_scatter = cap_mice.loc[:, ["Mouse ID", "Weight (g)", "Tumor Volume (mm3)"]]
average_cap_mice = pd.DataFrame(cap_mice_scatter.groupby(["Mouse ID", "Weight (g)"])["Tumor Volume (mm3)"].mean()).reset_index()
average_cap_mice = average_cap_mice.rename(columns={"Tumor Volume (mm3)": "Average Tumor Volume"})
average_cap_mice = average_cap_mice.set_index('Mouse ID')

average_cap_mice.plot(kind="scatter", x="Weight (g)", y="Average Tumor Volume", grid=True, figsize=(4,4),
              title="Mouse Weight vs. Average Tumor Volume for Capomulin")

```

# Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
weight = average_cap_mice["Weight (g)"]
tumor_volume = average_cap_mice["Average Tumor Volume"]

correlation = st.pearsonr(weight,tumor_volume)
print(f"The correlation between both factors is {round(correlation[0],2)}")
```

```python
x_values = average_cap_mice["Weight (g)"]
y_values = average_cap_mice["Average Tumor Volume"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = weight * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(weight, tumor_volume)
plt.plot(weight,regress_values, "r-")
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.xlabel('Mouse Weight')
plt.ylabel('Average Tumor Volume')
```
